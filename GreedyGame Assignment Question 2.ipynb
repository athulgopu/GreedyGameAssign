```python
#Importing the necessary libraries
import numpy as np
import pandas as pd
```

```python
# Importing the necessary data
xls = pd.ExcelFile('Q2_rewards details.xlsx')
user_signup = pd.read_excel(xls, 'Q2_users signup')
print('user_signup',user_signup.shape)
user_signup.head()
```

```python
user_offer = pd.read_excel(xls, 'Q2_User offer data')
print('user_offer',user_offer.shape)
user_offer.head()
```

```python
user_offer_comp = pd.read_excel(xls, 'Q2_User offer completion data')
print('user_offer_comp',user_offer_comp.shape)
user_offer_comp.head()
```

```python
rewards = pd.read_excel(xls, 'Q2_rewards details')
print('rewards',rewards.shape)
rewards.head()
```

```python
# merging to get the user_id,app_id and offer_id together
df=pd.merge(user_offer,user_signup[['user_id','app_id']],on='user_id',how='left')
print('df',df.shape)
df.head()
```

```python
# Number of offers started in each app
off_start=df.groupby('app_id')['user_id'].count().reset_index().rename(columns={'user_id':'offers_started'})
off_start
```

```python
# Number of offers completed in each app
off_comp=df[df.completed_at.notnull()].groupby('app_id')['user_id'].count().reset_index().rename(columns={'user_id':'offers_completed'})
off_comp
```

```python
# Completion% in each app
off_df=pd.merge(off_start,off_comp,on='app_id',how='left')
off_df['completion%']=(off_df['offers_completed']*100/off_df['offers_started']).round(2)
off_df
```

#### * Its clear from the above data that sikka pro is better than sikka in terms of completion rate.

```python
# merging to get the total revenue and profit
rev_df=pd.merge(user_offer_comp,rewards[['reward_id','total_payout_in_paise','total_revenue_in_paise']],on='reward_id',how='left')
print('rev_df',rev_df.shape)
rev_df.head()
```

```python
rev_df['total_profit_in_paise']=rev_df['total_revenue_in_paise']-rev_df['total_payout_in_paise']
rev_df.head()
```

```python
rev_df.groupby('app_id')['total_payout_in_paise','total_revenue_in_paise','total_profit_in_paise'].sum()
```

#### * Eventhough the completion rate was higher for sikka pro the total profit generated was slightly lesser than that generated from sikka.One reason for this could be due to users completing less number of high rewarding offers in sikka which returns more profit and users completing high number of low rewarding offers where the profit return is also lesser.
